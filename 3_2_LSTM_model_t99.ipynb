```python
def make_dataset(data, label, window_size = 39):
    feature_list = []
    label_list = []
    i = 0
    for t in type:
        for i in range(len(data)-window_size):
            feature_list.append(np.array(data[t].iloc[i:i+window_size]))
            label_list.append(np.array(label[t].iloc[i+window_size]))
       
    return np.array(feature_list), np.array(label_list)
```

```python
feature_cols = type
#label_cols = ['type 39']

train_feature = train[feature_cols]
train_label = train[feature_cols].iloc[-1]

train_feature, train_label = make_dataset(train_feature, train_label)

from sklearn.model_selection import train_test_split
x_train, x_valid, y_train, y_valid = train_test_split(train_feature, train_label, test_size=0.2)

# x_train.shape, x_valid.shape
```
