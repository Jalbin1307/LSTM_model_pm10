```python
f = open('list.csv', 'r')
rdr = csv.reader(f)
num = 0
b = []
for line in rdr:
    num = num + 1
    if len(line[0].split(',')) != 40:
        print(num, len(line[0].split(',')))
    else:
        b.append(line[0].split(','))    
f.close()
```

```python
f = open('dt_v.csv','w', newline='')
writer = csv.writer(f)
writer.writerow(["type %d"%i for i in range(len(b))])
for i in range(len(b[0])):
    writer.writerow([b[j][i] for j in range(len(b))])
f.close()
```
